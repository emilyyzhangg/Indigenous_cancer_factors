```python
# importing necessary packages
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# uploading CSV file
cancer_risks = pd.read_csv('Indigenous_cancer_risk.csv')

cancer_risks
```

```python
# turning 95CI values into arrays of tuples for each cancer type
colorectal_eb_df = pd.DataFrame().assign(lower=cancer_risks['Colorectal 95CI LOWER'], upper=cancer_risks['Colorectal 95CI UPPER'])
colorectal_eb = list(colorectal_eb_df.itertuples(index=False, name=None))

lung_eb_df = pd.DataFrame().assign(lower=cancer_risks['Lung 95CI LOWER'], upper=cancer_risks['Lung 95CI UPPER'])
lung_eb = list(lung_eb_df.itertuples(index=False, name=None))

breast_eb_df = pd.DataFrame().assign(lower=cancer_risks['Breast 95CI LOWER'], upper=cancer_risks['Breast 95CI UPPER'])
breast_eb = list(breast_eb_df.itertuples(index=False, name=None))

prostate_eb_df = pd.DataFrame().assign(lower=cancer_risks['Prostate 95CI LOWER'], upper=cancer_risks['Prostate 95CI UPPER'])
prostate_eb = list(prostate_eb_df.itertuples(index=False, name=None))
```

```python
# turning province column into an array
provinces = cancer_risks['Province'].to_numpy()

# create data
x = np.arange(9)
y1 = cancer_risks['Colorectal ASIR']
y2 = cancer_risks['Lung ASIR']
y3 = cancer_risks['Breast ASIR']
y4 = cancer_risks['Prostate ASIR']
width = 0.2
  
# plot data in grouped manner of bar type
plt.bar(x-0.3, y1, width, color='blue')
plt.bar(x-0.1, y2, width, color='orange')
plt.bar(x+0.1, y3, width, color='green')
plt.bar(x+0.3, y4, width, color='red')
plt.xticks(x, provinces, rotation='vertical')
plt.xlabel("Province")
plt.ylabel("Relative Cancer Risk")
plt.legend(["Colorectal Cancer", "Lung Cancer", "Breast Cancer", "Prostate Cancer"])
plt.axhline(y=1, linestyle = 'dashed')

plt.show()
```
